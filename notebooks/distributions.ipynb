```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
tb_data = pd.read_csv('../data/../data/train.csv', index_col='id')
tb_data
```

```python
warnings.filterwarnings(action='ignore')

# f0 - f24
col_1 = tb_data.columns[:25]
figure_1, axes_1 = plt.subplots(nrows=5, ncols=5, figsize=[24, 24], dpi=300)
axes_1 = axes_1.ravel()

for i in range(len(axes_1)):
    sns.distplot(x=tb_data[col_1[i]], ax=axes_1[i])
    axes_1[i].set_xlabel(f'{col_1[i]}')

plt.show()
```

```python
# f25 - f50
col_2 = tb_data.columns[25:50]
figure_2, axes_2 = plt.subplots(nrows=5, ncols=5, figsize=[24, 24], dpi=300)
axes_2 = axes_2.ravel()

for i in range(len(axes_2)):
    sns.distplot(x=tb_data[col_2[i]], ax=axes_2[i])
    axes_2[i].set_xlabel(f'{col_2[i]}')

plt.show()
```
